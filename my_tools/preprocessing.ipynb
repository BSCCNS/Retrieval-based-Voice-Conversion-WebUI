```python
# Target parameters
TARGET_SAMPLE_RATE = 48000
TARGET_BIT_DEPTH = "pcm_s32le" # "pcm_s24le"  # PCM 24-bit
out_dir = f'{root}/48k'

try:
    os.mkdir(out_dir)
except:
    pass
```

```python
# Process each file
for file_path in files:
    # Read file properties
    with sf.SoundFile(file_path) as f:
        sample_rate = f.samplerate
        bit_depth = f.subtype

    # Load the audio with pydub
    audio = AudioSegment.from_wav(file_path)

    
    audio = audio.set_frame_rate(TARGET_SAMPLE_RATE)

    target = file_path.split(root)[-1].replace('.wav', '_48k.wav')
    output_file = f'{out_dir}/{target}' 

    audio.export(output_file, 
                 format= "wav", 
                 parameters = ["-acodec", TARGET_BIT_DEPTH])
```

```python
root = '/Users/tomasandrade/Documents/BSC/ICHOIR/organos/dataset/flat/48k/mono'
```

```python
files = glob.glob(f'{root}/*.wav', recursive=False)
files
```

```python
out_dir = f'{root}/spliced'
os.mkdir(out_dir)
```

```python
from pydub.utils import make_chunks
```

```python
chunk_length_ms = 1000*4 # pydub calculates in millisec

for file in files:
    print(f'------- Working on file {file}')

    myaudio = AudioSegment.from_file(file , "wav") 
    chunks = make_chunks(myaudio, chunk_length_ms) 
    name = file.split('/')[-1].split('.wav')[-2].replace(' ', '_')

    print(name)

    for i, chunk in enumerate(chunks):
        chunk_name = f"{name}_{i}.wav"
        print(f"exporting {chunk_name}")
        chunk.export(f'{out_dir}/{chunk_name}', format="wav")
```
