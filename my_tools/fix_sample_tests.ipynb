```python
# Target parameters
TARGET_SAMPLE_RATE = 48000
TARGET_BIT_DEPTH = "pcm_s24le"  # PCM 24-bit
out_dir = f'{root}/48k'

try:
    os.mkdir(out_dir)
except:
    pass

# Process each file
for file_path in files:
    # Read file properties
    with sf.SoundFile(file_path) as f:
        sample_rate = f.samplerate
        bit_depth = f.subtype

    # Load the audio with pydub
    audio = AudioSegment.from_wav(file_path)

    
    audio = audio.set_frame_rate(TARGET_SAMPLE_RATE)

    target = file_path.split(root)[-1].replace('.wav', '_48k.wav')
    output_file = f'{out_dir}/{target}' 

    audio.export(output_file, 
                 format= "wav", 
                 parameters = ["-acodec", TARGET_BIT_DEPTH])
    
    # if bit_depth != "PCM_24":  # Convert FLOAT or other formats to PCM_24

    #     target = file_path.split(root)[-1].replace('.wav', '_48k.wav')
    #     output_file = f'{out_dir}/{target}' 

    #     audio.export(output_file, format="wav", parameters=["-acodec", TARGET_BIT_DEPTH])
    #     print(f"Converted: {file_path} → {output_file} (48kHz, PCM_24)")
    # else:
    #     print(f"Already in correct format: {file_path}")
    #     shutil.copy(file_path, output_file)
```

```python
for file in files:
    print(file)
    audio = AudioSegment.from_wav(file)
    mono_audios = audio.split_to_mono()

    target_left = file.split(root)[-1].replace('.wav', '_left.wav')
    output_left_file = f'{out_dir}/{target_left}'

    target_right = file.split(root)[-1].replace('.wav', '_right.wav')
    output_right_file = f'{out_dir}/{target_right}'

    mono_audios[0].export(output_left_file, format="wav", parameters=["-acodec", "pcm_s24le"])
    mono_audios[1].export(output_right_file, format="wav", parameters=["-acodec", "pcm_s24le"])
```

```python
from pydub.utils import make_chunks
```

```python
chunk_length_ms = 1000*24 # pydub calculates in millisec

files = glob.glob(f'{root}/*.wav', recursive=False)
print(files)

for file in files:

    print(f'------- Working on file {file}')

    myaudio = AudioSegment.from_file(file , "wav") 
    chunks = make_chunks(myaudio, chunk_length_ms) 
    name = file.split('/')[-1].split('.wav')[-2].replace(' ', '_')

    print(name)

    for i, chunk in enumerate(chunks):
        chunk_name = f"{name}_{i}.wav"
        print(f"exporting {chunk_name}")
        chunk.export(f'{out_dir}/{chunk_name}', format="wav")
```
